```python

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Seleccionar las columnas numéricas para PCA (o las que deseas visualizar)
datos_num = datos.select_dtypes(include=[float, int])

# Reducir a 2 dimensiones para la visualización
pca = PCA(n_components=2)
pca_result = pca.fit_transform(datos_num)

# Graficar los resultados de PCA
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=datos['Cluster'], cmap='viridis')

# Etiquetar los ejes con los componentes principales
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Visualización de Clustering con PCA')

plt.show()
```
